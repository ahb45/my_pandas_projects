### Baharestan Pockets : use bahar_sandali_All.xlsx
#### result is : bahar_prep1.xlsx

```python
import pandas as pd
import os
import shutil
from openpyxl import Workbook
from openpyxl.styles import Alignment
from openpyxl.utils import get_column_letter

data = pd.read_excel('bahar_sandali_All.xlsx')  
df = pd.DataFrame(data)  


df['tedad'] = 1

# گروه‌بندی و جمع‌آوری داده‌ها
result = df.groupby(['makan','code']).agg(
    tedad2=('tedad', 'sum'),
    sandaliFirst=('sandali', 'min'),
    sandaliEnd=('sandali', 'max'),
    tarikh=('tarikh','first'),
    saat=('saat','first'),
    #makan=('makan','first'),
    noe=('noe','first'),
    dars=('dars','first')

    
).reset_index()

#result = result.merge(df[['code', 'dars','noe','tarikh','saat']], on='code', how='left')
result = result.sort_values(by=['tarikh', 'saat','makan'])

result.to_excel('bahar_prep1.xlsx', index=False)

print("DataFrame successfully saved to bahar_prep1.xlsx")
```

### Baharestan Generate Pockets : use bahar_prep1.xlsx

```python
data = pd.read_excel('bahar_prep1.xlsx')  
df = pd.DataFrame(data)  

# تغییر ترتیب ستون‌ها
df = df[['tedad2','makan','noe','sandaliEnd','sandaliFirst','saat','tarikh','code','dars']]

unique_values_tarikh = df['tarikh'].unique()
unique_values_saat   = df['saat'].unique()
unique_values_makan  = df['makan'].unique()
if os.path.exists('pages'):
    shutil.rmtree('pages')
os.makedirs('pages')
k = 1

for ta in unique_values_tarikh:
    for sa in unique_values_saat: 
        for ma in unique_values_makan:
          str1 = 'bahar_report' + str(k) + '.xlsx'
          result = df[(df['tarikh']== ta) & (df['saat']== sa) & (df['makan']== ma)]
          #result = result.sort_values(by=['makan'])
          #result.to_excel('pages//'+str1, index=False)
          # نوشتن دیتافریم به فایل اکسل
          if len(result) > 0:
            with pd.ExcelWriter('pages//'+str1, engine='openpyxl') as writer:
                result.to_excel(writer, index=False, sheet_name='Sheet1')
      

                # دسترسی به شیء workbook و worksheet
                workbook = writer.book
                worksheet = writer.sheets['Sheet1']
                worksheet.page_setup.orientation = worksheet.ORIENTATION_LANDSCAPE

                # تنظیم عرض ستون‌ها
                worksheet.column_dimensions['A'].width = 80 / 7.5  # تبدیل پیکسل به واحد اکسل
                worksheet.column_dimensions['B'].width = 80 / 7.5 
                worksheet.column_dimensions['C'].width = 60 / 7.5
                worksheet.column_dimensions['D'].width = 50 / 7.5
                worksheet.column_dimensions['E'].width = 50 / 7.5
                worksheet.column_dimensions['F'].width = 80 / 7.5
                worksheet.column_dimensions['G'].width = 80 / 7.5
                worksheet.column_dimensions['H'].width = 80 / 7.5
                worksheet.column_dimensions['I'].width = 200 / 7.5
                
                columns_to_center = [1,2,4,5,6,7,8]
                for col in columns_to_center:
                  for row in range(2, len(df) + 2):  # از ردیف 2 شروع می‌کنیم چون ردیف 1 سرستون‌هاست
                    cell = worksheet.cell(row=row, column=col)   
                    cell.alignment = Alignment(horizontal='center')   
              
          k = k+1
```

```python
data = pd.read_excel('bahar_sandali_All.xlsx')  
df = pd.DataFrame(data)  

unique_values_tarikh = df['tarikh'].unique()
sorted_tarikh = sorted(unique_values_tarikh)
unique_values_saat   = df['saat'].unique()
sorted_saat = sorted(unique_values_saat)
if os.path.exists('site_pages'):
    shutil.rmtree('site_pages')
os.makedirs('site_pages')
k = 1

for ta in sorted_tarikh:
    for sa in sorted_saat: 
        str1 = 'sandali' + str(k) + '.xlsx'
        result = df[(df['tarikh']== ta) & (df['saat']== sa)]
        result = result.sort_values(by=['tarikh','saat','code'])
        #result.to_excel('site_pages//'+str1, index=False)
        with pd.ExcelWriter('site_pages//'+str1, engine='openpyxl') as writer:
                result.to_excel(writer, index=False, sheet_name='Sheet1')
      

                # دسترسی به شیء workbook و worksheet
                workbook = writer.book
                worksheet = writer.sheets['Sheet1']
                #worksheet.page_setup.orientation = worksheet.ORIENTATION_LANDSCAPE

                # تنظیم عرض ستون‌ها
                worksheet.column_dimensions['A'].width = 100 / 7.5  # تبدیل پیکسل به واحد اکسل
                worksheet.column_dimensions['B'].width = 70 / 7.5 
                worksheet.column_dimensions['C'].width = 90 / 7.5
                worksheet.column_dimensions['D'].width = 70 / 7.5
                worksheet.column_dimensions['E'].width = 230 / 7.5
                worksheet.column_dimensions['F'].width = 90 / 7.5
                worksheet.column_dimensions['G'].width = 90 / 7.5
                worksheet.column_dimensions['H'].width = 80 / 7.5
                worksheet.column_dimensions['I'].width = 80 / 7.5
                worksheet.column_dimensions['J'].width = 120 / 7.5
                
                columns_to_center = [6,7,8]
                for col in columns_to_center:
                  for row in range(2, len(df) + 2):  # از ردیف 2 شروع می‌کنیم چون ردیف 1 سرستون‌هاست
                    cell = worksheet.cell(row=row, column=col)   
                    cell.alignment = Alignment(horizontal='center')   
        k = k+1
```

```python
import pandas as pd
from docx import Document
from docx.enum.section import WD_ORIENT
from docx.shared import Cm
from docx.enum.text import WD_ALIGN_PARAGRAPH

# تعداد فایل‌ها
num_files = 798

# ایجاد یک فایل ورد جدید
doc = Document()
doc.add_heading('محتوای فایل‌های اکسل', level=1)

# تنظیم جهت صفحه به landscape
section = doc.sections[-1]
section.orientation = WD_ORIENT.LANDSCAPE
section.page_width, section.page_height = section.page_height, section.page_width

# پردازش فایل‌های اکسل
for i in range(1, num_files + 1):
    # تولید نام فایل
    file = f"pages//bahar_report{i}.xlsx"

    try:
        # خواندن فایل اکسل
        df = pd.read_excel(file)
    except FileNotFoundError:
        print(f"فایل پیدا نشد: {file}")
        continue  # به فایل بعدی بروید
    
    # اضافه کردن صفحه جدید برای هر فایل
    if i > 1:  # صفحه جدید به جز برای فایل اول
        doc.add_page_break()

    # اضافه کردن عنوان برای هر فایل
    #doc.add_heading(f'فایل {i}: ', level=2)
    
    # اضافه کردن محتوای جدول اکسل به ورد
    table = doc.add_table(rows=1, cols=len(df.columns))
    table.style = 'Table Grid'
    
    # تنظیم عرض ستون اول
    table.columns[8].width = Cm(15.08)  # 5.08 cm = 200 pixels

    # اضافه کردن عنوان ستون‌ها
    hdr_cells = table.rows[0].cells
    for j, column_name in enumerate(df.columns):
        hdr_cells[j].text = str(column_name)
    
    # اضافه کردن داده‌ها و تنظیم چینش ستون‌ها
    for _, row in df.iterrows():
        row_cells = table.add_row().cells
        for j, value in enumerate(row):
            cell = row_cells[j]
            cell.text = str(value)

            # تنظیم چینش ستون‌ها
            if j == 0:  # ستون اول (وسط‌چین)
                cell.paragraphs[0].paragraph_format.alignment = WD_ALIGN_PARAGRAPH.CENTER
            elif j == 8:  # ستون نهم (راست‌چین)
                cell.paragraphs[0].paragraph_format.alignment = WD_ALIGN_PARAGRAPH.RIGHT

# ذخیره فایل ورد
output_file = "output.docx"
doc.save(output_file)
print(f"فایل ورد با موفقیت ذخیره شد: {output_file}")




```
